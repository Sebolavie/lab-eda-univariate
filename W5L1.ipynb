```python
import pandas as pd

# Load the dataset
file_path = 'Documents/GitHub/lab-eda-univariate/amz_dataset.csv'
amazon_df = pd.read_csv(file_path)
```

```python
amazon_df.columns
```

```python
amazon_df.head ()

```

```python
# Generate frequency table for product categories
category_counts = amazon_df['category'].value_counts()
```

```python
print(category_counts)
```

```python
top_5_categories = category_counts.head(5)
print(top_5_categories)
```

```python
import matplotlib.pyplot as plt

# Bar chart for the distribution of products across all categories
plt.figure(figsize=(12, 8))
category_counts.plot(kind='bar')
plt.title('Distribution of Products Across Different Categories')
plt.xlabel('Product Category')
plt.ylabel('Number of Listings')
plt.show()
```

```python
# Pie chart for the top 5 categories
plt.figure(figsize=(10, 7))
top_5_categories.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportions of Top 5 Product Categories')
plt.ylabel('')
plt.show()
```

```python
# Calculate mean, median, and mode for product prices
mean_price = amazon_df['price'].mean()
median_price = amazon_df['price'].median()
mode_price = amazon_df['price'].mode()[0]

(mean_price, median_price, mode_price)
```

```python
# Calculate variance, standard deviation, range, and IQR for product prices
variance_price = amazon_df['price'].var()
std_dev_price = amazon_df['price'].std()
range_price = amazon_df['price'].max() - amazon_df['price'].min()
q1 = amazon_df['price'].quantile(0.25)
q3 = amazon_df['price'].quantile(0.75)
iqr_price = q3 - q1

(variance_price, std_dev_price, range_price, iqr_price)
```

```python
# Histogram for the distribution of product prices
plt.figure(figsize=(10, 6))
plt.hist(amazon_df['price'], bins=50, edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Box plot for product prices
plt.figure(figsize=(10, 6))
plt.boxplot(amazon_df['price'], vert=False)
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Filter the dataset for top 5 categories
top_5_df = amazon_df[amazon_df['category'].isin(top_5_categories)]
```

```python
# Histogram for the distribution of product prices
plt.figure(figsize=(10, 6))
plt.hist(top_5_df['price'], bins=50, edgecolor='black')
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for product prices in top 5 categories
plt.figure(figsize=(10, 6))
plt.boxplot(top_5_df['price'], vert=False)
plt.title('Box Plot of Product Prices in Top 5 Categories')
plt.xlabel('Price')
plt.show()
```

```python
#challenge3

# Calculate mean, median, and mode for product ratings
mean_rating = amazon_df['reviews'].mean()
median_rating = amazon_df['reviews'].median()
mode_rating = amazon_df['reviews'].mode()[0]

(mean_rating, median_rating, mode_rating)
```

```python
# Calculate variance, standard deviation, and IQR for product ratings
variance_rating = amazon_df['reviews'].var()
std_dev_rating = amazon_df['reviews'].std()
q1 = amazon_df['reviews'].quantile(0.25)
q3 = amazon_df['reviews'].quantile(0.75)
iqr_rating = q3 - q1

(variance_rating, std_dev_rating, iqr_rating)
```

```python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis for product ratings
skewness_rating = skew(amazon_df['reviews'].dropna())
kurtosis_rating = kurtosis(amazon_df['reviews'].dropna())

(skewness_rating, kurtosis_rating)
```

```python
# Histogram for the distribution of product ratings
plt.figure(figsize=(10, 6))
plt.hist(amazon_df['reviews'], bins=50, edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
